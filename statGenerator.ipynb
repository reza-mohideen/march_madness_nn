```python
year = 2005
TEAM_DIR = 'D:/PycharmProjects/marchMadness/teams/' + str(year) + '/'
#make list of conferences in one year
urlLeague = 'https://www.sports-reference.com/cbb/seasons/' + str(year) + '.html'
conferences = pd.read_html(urlLeague)
conferenceList = []
for conf in conferences[0]['Conference']:
    if conf == 'Atlantic Coast Conference':
        conferenceList.append('acc')
    elif conf == 'Southeastern Conference':
        conferenceList.append('sec')
    elif conf == 'Pacific-12 Conference':
        conferenceList.append('pac-12')
    elif conf == 'Pacific-10 Conference':
        conferenceList.append('pac-10')
    elif conf == 'American Athletic Conference':
            conferenceList.append('aac')
    elif conf == 'Mountain West Conference':
         conferenceList.append('mwc')
    elif conf == 'Missouri Valley Conference':
        conferenceList.append('mvc')
    elif conf == 'West Coast Conference':
        conferenceList.append('wcc')
    elif conf == 'Conference USA':
        conferenceList.append('cusa')
    elif conf == 'Mid-American Conference':
        conferenceList.append('mac')
    elif conf == 'Colonial Athletic Association':
        conferenceList.append('colonial')
    elif conf == 'Western Athletic Conference':
        conferenceList.append('wac')
    elif conf == 'Colonial Athletic Association':
        conferenceList.append('colonial')
    elif conf == 'Ivy Group':
        conferenceList.append('ivy')
    elif conf == 'Metro Atlantic Athletic Conference':
        conferenceList.append('maac')
    elif conf == 'Ohio Valley Conference':
        conferenceList.append('ovc')
    elif conf == 'Southwest Athletic Conference':
        conferenceList.append('swac')
    elif conf == 'Mid-Eastern Athletic Conference':
        conferenceList.append('meac')
    else: 
        conferenceList.append(conf.replace('Conference','').replace('League','').lower().rstrip().replace(' ','-'))
print(conferenceList)
    
```

```python
from dateutil.parser import parse
import os
from tqdm import tqdm
```

```python
winCount = 0
lossCount = 0
year = 2005
print(year)
TEAM_DIR = 'D:/PycharmProjects/marchMadness/teams/' + str(year)
teamFiles = []
teams = []
for team in os.listdir(TEAM_DIR):
    if team.split('.')[1] == 'csv':
        teamFiles.append(team)
        teams.append(team.split('.')[0])
#print(teamFiles)
for file in (teamFiles):
    #clear_output(wait = True)
    f = TEAM_DIR + '/' + file
    rows = 0
    with open(f) as csvfile:
        readCSV = csv.reader(csvfile, delimiter=',')
        cGame = []
        cDates = []
        cOpponent = []
        cTm = []
        cOpp = []
        cPdf = []
        cPf = []
        cPa = []
        for row in readCSV:
            if row[0] != 'G':
                cGame.append(int(row[0]))
                dt = parse(row[1])
                cDates.append(dt)
                cOpponent.append(row[2])
                cTm.append(int(row[3]))
                cOpp.append(int(row[4]))
                cPdf.append((row[5]))
                cPf.append((row[6]))
                cPa.append((row[7]))
                rows += 1

cTeam = {'game': cGame,'date':cDates,'opponent': cOpponent,'tm': cTm,'opp': cOpp,'pdf': cPdf,'pf': cPf,'pa': cPa}
print(cTeam)     
TRAIN_DIR = 'D:/PycharmProjects/marchMadness/train'
        
for game in range(1,rows):
    opponent = cTeam['opponent'][game]
    opponent = ''.join([i for i in opponent if not i.isdigit()])
    opponent = opponent.rstrip()
    date = cTeam['date'][game]
    score = [cTeam['tm'][game],cTeam['opp'][game]]
    if score[0] > score[1]:
        label = 'win'
    else:
        label = 'loss'
    score = [str(cTeam['tm'][game]),str(cTeam['opp'][game])]
    cStats = [cTeam['pdf'][game-1],cTeam['pf'][game-1],cTeam['pa'][game-1]]
        #print(cStats)
    try:
        with open(opponent + '.csv') as csvfile:
            readCSV = csv.reader(csvfile, delimiter=',')
            for row in readCSV:
                if row[0] != 'G':
                    if parse(row[1]) == date:
                        oppGame = int(row[0])
                        oppGame -= 1
                        oppGame = str(oppGame)
                        with open(opponent + '.csv') as oppfile:
                            oppCSV = csv.reader(oppfile, delimiter=',')
                            for oppRow in oppCSV:
                                if oppRow[0] == oppGame:
                                    oppStats = [oppRow[5],oppRow[6],oppRow[7]]
    except FileNotFoundError:
        continue

                            #print(oppStats)
    finalFile = [cStats,oppStats,score]
    print(finalFile)
    if label == 'win':
        with open(TRAIN_DIR + '/' + label + '.' + str(winCount) + '.' + f.split('.')[0] + '.' + str(game+1)+ '.' + str(year) + '.csv', 'w') as csvFile:
            writer = csv.writer(csvFile, lineterminator = '\n')
            writer.writerows(finalFile)
            winCount += 1
        csvFile.close()
                
    elif label == 'loss':
        with open(TRAIN_DIR + '/' + label + '.' + str(lossCount) + '.' + f.split('.')[0] + '.' + str(game+1)+ '.' + str(year) + '.csv', 'w') as csvFile:
            writer = csv.writer(csvFile, lineterminator = '\n')
            writer.writerows(finalFile)
            lossCount += 1
        csvFile.close()
                
        
```
